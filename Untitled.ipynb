```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data.csv')
cols = ['Q13_1: Sustained Attention to Response Task (SART): https://www.psytoolkit.org/experiment-library/experiment_sart.html - Number Go trials',
        'Q13_2: Sustained Attention to Response Task (SART): https://www.psytoolkit.org/experiment-library/experiment_sart.html - Number Go mistakes',
        'Q13_3: Sustained Attention to Response Task (SART): https://www.psytoolkit.org/experiment-library/experiment_sart.html - Go mistakes (%)',
        'Q13_4: Sustained Attention to Response Task (SART): https://www.psytoolkit.org/experiment-library/experiment_sart.html - Number No Go trials',
        'Q13_5: Sustained Attention to Response Task (SART): https://www.psytoolkit.org/experiment-library/experiment_sart.html - Number No Go mistakes',
        'Q13_6: Sustained Attention to Response Task (SART): https://www.psytoolkit.org/experiment-library/experiment_sart.html - No Go mistakes (%)',
        'Q26_1: Autism Spectrum Quotient: https://www.psytoolkit.org/cgi-bin/3.4.4/survey?s=cRqHQ - Score:',
       ]

cols2 = ['Q9_1: Stroop taak: https://www.psytoolkit.org/experiment-library/experiment_stroop.html - Congruent',
        'Q9_2: Stroop taak: https://www.psytoolkit.org/experiment-library/experiment_stroop.html - Incongruent',
        'Q9_3: Stroop taak: https://www.psytoolkit.org/experiment-library/experiment_stroop.html - Stroop effect',
        'Q26_1: Autism Spectrum Quotient: https://www.psytoolkit.org/cgi-bin/3.4.4/survey?s=cRqHQ - Score:']
```

```python
# data = data[cols].dropna()
data1 = data[cols].dropna()
data2 = data[cols2].dropna()
```

```python
cols_renamed = {i: f'SART - {i.split("-")[-1]}' for i in cols[:-1]}
cols2_renamed = {i: i.split("-")[-1][1:] for i in cols2[:-1]}

data1 = data1.rename(columns = cols_renamed)
data1 = data1.rename(columns = {'Q26_1: Autism Spectrum Quotient: https://www.psytoolkit.org/cgi-bin/3.4.4/survey?s=cRqHQ - Score:': 'ASQ score:'})

data2 = data2.rename(columns = cols2_renamed)
data2 = data2.rename(columns = {'Q26_1: Autism Spectrum Quotient: https://www.psytoolkit.org/cgi-bin/3.4.4/survey?s=cRqHQ - Score:': 'ASQ score:'})
display(data2)

def fix_data(value):
    if isinstance(value, int):
        return value
    elif isinstance(value, str):
        return int(value.rstrip('ms'))
    else:
        return int(value)
```

```python
data2['Stroop effect'] = data2['Stroop effect'].map(fix_data)
```

```python
def fix_percentage(column):
        if isinstance(column, float):
            return column
        elif '%' in column:
                return float(column.rstrip('%'))
        else:
                return float(column)
            
data1['SART -  No Go mistakes (%)'] = data1['SART -  No Go mistakes (%)'].apply(fix_percentage)
data1['SART -  Go mistakes (%)'] = data1['SART -  Go mistakes (%)'].apply(fix_percentage)

display(data1.sort_values(by='SART -  Number Go trials'))
```
